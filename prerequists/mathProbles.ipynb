4.LCM of a number

```python
def lcm(a,b):
	maxi = max(a,b)
	minm = min(a,b)
	for i in range(maxi, a * b + 1, maxi):
		if i % minm == 0:
			return i
	return a * b

if __name__ == '__main__':
	a = 11
	b = 5
	print(lcm(a,b))
```

6.GCD OF TWO NUMBERS

```python
def gcd(a,b):
  minValue = min(a,b)
  while minValue > 0:
    if (a % minValue == 0 and b % minValue == 0):
      break
    minValue -= 1
  return minValue

a = 20
b = 28
print(gcd(a,b))  
```

8.SecondLargest NUmber from array or return -1 if not found

```python
def secondLargest(arr):
  n = len(arr)
  arr.sort()
  for i in range(n-2, -1, -1):
    if arr[i] != arr[n-1]:
      return arr[i]
  return -1
      
arr = [10,20,30,2,4,5.10,50]
print(secondLargest(arr))
arr2 = [10,10,10]
print(secondLargest(arr2))

arr1 = [20,10,1,2,3,50]
print(secondLargest(arr1))
```
