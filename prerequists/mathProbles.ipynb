1.Positive or Negative Number

```python
num = 10
if num > 0:
	print("Positive")
elif num < 0:
	print("negative")
else:
	print("zero")
```

```python
num1 = 10
if num1 >= 0:
	if num1 > 0:
		print("positive")
	else:
		print("zero")
else:
	print("Negative")
```

```python
num2 = -1

print("Positive" if num2 >= 0 else "Negative" )
```

2.Even or ODd

```python
num = 1
if num % 2 == 0:
  print("even")
elif num % 2 != 0:
  print("odd")
```

```python
num = 1
print("Even" if num % 2 == 0 else "Odd")
```

```python
def isEven(num):
  return not num & 1

num = 10
if isEven(num):
  print("Even")
else:
  print("odd")
```

3.Sum Of N Natural Numbers

```python
sum = 0
num = 6
for i in range(1,num + 1):
  sum = sum + i

print(sum)
```

```python
num = 5
sum = int(num * ((num + 1) / 2))
print(sum)
```

4.LCM of a number

```python
def lcm(a,b):
	maxi = max(a,b)
	minm = min(a,b)
	for i in range(maxi, a * b + 1, maxi):
		if i % minm == 0:
			return i
	return a * b

if __name__ == '__main__':
	a = 11
	b = 5
	print(lcm(a,b))
```
