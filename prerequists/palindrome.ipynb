Palindrome:

Method1: simple iteration

```python
number = 12321
temp = number
reverse = 0
while temp > 0:
  reverse = ( reverse * 10 ) + temp % 10 
  temp = temp // 10

if reverse == number:
  print("plaindrome")
else:
  print("not palindrome")
```

Method2: String Slicing

```python
number = '123454321'
if number == number[::-1]:
  print(number,":Number is palindrome")
else:
  print("not palindrome")
```

Method3:Using Character matching

```python
number = '12321'
length = len(number)
for i in range(0,length):
  length = length - 1 
  if number[i] == number[length]:
    
    print('palindrome')
  else:
    print("not palindrome")
```

DSA LEVEL : TWO POINTER APPROACH

```python
number = "1245421" 
left = 0
right = len(number) - 1
is_palindrome = True
while left < right:
  if number[left] != number[right]:
    is_palindrome = False
    break
  left += 1
  right -= 1

if is_palindrome:
  print("palindrome")
else:
  print("not Palindrome")
```
