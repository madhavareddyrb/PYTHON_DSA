## SUM OF DIGITS

```python
number = 1234
sum = 0
for i in str(number):
  sum += int(number)  % 10
  number = number / 10
print(sum) 
```

Method 0: Using String Character Extraction

```python
number = '12345'
sum = 0
for i in number:
  sum += int(i)

print(sum)
```

Method 1: Using Brute Force

```python
num = 12345
sum = 0

while num != 0:
  digit = int(num % 10 )
  sum += digit
  num = num // 10

print(sum)
```
