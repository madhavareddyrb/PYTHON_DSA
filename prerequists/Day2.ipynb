Prime Numbers in Range

```python
n = 2
m = 10
count = 0
lis = []
for i in range(n,m):
  count = 0
  for j in range(2,i):
    if i % j == 0:
      count = 1
      break
  if count == 0:
     lis.append(i)

print(lis)
```

```python
import math
n = 2
m = 10
primes = []
for i in range(n,m):
  if i < 2:
    continue
  is_prime = True
  for j in range(2,int(math.sqrt(i))+1):
    if i % j == 0:
      is_prime  = False
      break
    
  if is_prime:
    primes.append(i)
    
print(primes)
```
