## Armstrong Number:
example = 371 = 3** 3 + 7** 3 + 1** 3 = 371 armstrong number

### A number which is equal to power of length for each digit should be same

Method 1: Iteration

```python
num = '1634'
leng = len(num)
sum = 0

for i in range(leng):
  sum += (int(num[i]) ** leng)

## print(sum)

if sum == int(num):
  print(num, "Its a armstrong number")
else:
  print(num, "its Not an armstrong number")
```
