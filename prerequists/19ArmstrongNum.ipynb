## Armstrong Number:
example = 371 = 3** 3 + 7** 3 + 1** 3 = 371 armstrong number

### A number which is equal to power of length for each digit should be same

Method 1: Iteration

```python
num = '1634'
leng = len(num)
sum = 0

for i in range(leng):
  sum += (int(num[i]) ** leng)

## print(sum)

if sum == int(num):
  print(num, "Its a armstrong number")
else:
  print(num, "its Not an armstrong number")
```

The above method is changeing to str and makeing to int which takes more space and time so below we did using math without type casting for everytime

```python
num = 1634
num_len = len(str(num))
temp = num
digit_sum = 0
for i in range(num_len):
  digit = temp % 10
  digit_sum += digit ** num_len
  temp //= 10

if digit_sum == num:
  print(num, "Armstrong Number")
else:
  print(num, 'Its Not an armstrong number')
```
