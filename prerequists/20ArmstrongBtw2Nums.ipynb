### Find the Armstrong Numbers between Two Intervals 

```python
num1 = 10
num2 = 10000
for i in range(num1, num2):
  length = len(str(i))
  digit_sum = 0
  temp = i
  while temp > 0:
    digit = temp % 10
    digit_sum += digit ** length
    temp //= 10
  if digit_sum == i:
    print(i, end= ", ")

```
